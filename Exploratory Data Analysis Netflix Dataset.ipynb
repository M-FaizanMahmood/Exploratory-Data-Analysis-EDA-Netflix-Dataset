# Updating ratings based on different conditions

```python

# Condition 1: For 'Movies' listed_in with 'Unknown' rating
condition1 = (
    df_netflix['listed_in'].isin(['Movies']) & 
    df_netflix['rating'].isin(['Unknown'])
)
df_netflix.loc[condition1, 'rating'] = 'TV-MA'

# Condition 2: For 'Dramas' listed_in with 'Unknown' rating
condition2 = (
    df_netflix['listed_in'].isin(['Dramas']) & 
    df_netflix['rating'].isin(['Unknown'])
)
df_netflix.loc[condition2, 'rating'] = 'PG-13'

# Condition 3: For 'Anime Series, International TV Shows' listed_in with 'Unknown' rating
condition3 = (
    df_netflix['listed_in'].isin(['Anime Series, International TV Shows']) & 
    df_netflix['rating'].isin(['Unknown'])
)
df_netflix.loc[condition3, 'rating'] = 'TV-14'

# Condition 4: For 'Kids\' TV' or 'TV Comedies' listed_in with 'Unknown' rating
condition4 = (
    df_netflix['listed_in'].isin(['Kids\' TV', 'TV Comedies']) & 
    df_netflix['rating'].isin(['Unknown'])
)
df_netflix.loc[condition4, 'rating'] = 'TV-Y'
```

# Distribution of content across the different types (Movies/TV Shows) using 
matplotlib pie chart

```python
type_counts = df_netflix['type'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Content Across Different Types')
plt.axis('equal')

plt.show()
```
